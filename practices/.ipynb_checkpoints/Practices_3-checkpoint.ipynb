```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise 1: density peaks

```python
# import aggregation dataset
data = np.loadtxt("data/Aggregation.txt")
```

```python
# auxiliary functions for density peaks
def distance_matrix(data):
    fact_1 = np.tile(data.reshape((data.shape[0], 1, data.shape[1])), (1, data.shape[0], 1))
    fact_2 = np.tile(data.reshape((1, data.shape[0], data.shape[1])), (data.shape[0], 1, 1))
    return np.linalg.norm(fact_1-fact_2, axis=2)

def density_vector(dist_mat, dc):
    return np.sum(np.exp(- np.square(dist_mat/dc)), axis=1)

def radius_vector(dist_mat, dens_vec):
    comp_1 = np.tile(dens_vec.reshape((dens_vec.shape[0], 1)), (1, dens_vec.shape[0]))
    comp_2 = np.tile(dens_vec.reshape((1, dens_vec.shape[0])), (dens_vec.shape[0], 1))
    mask = comp_1 < comp_2
    return np.min(np.multiply(dist_mat, mask) + (np.multiply(np.logical_not(mask), np.max(dist_mat, axis=1))), axis=1)

def higher_dens_nn(dist_mat, dens_vec):
    comp_1 = np.tile(dens_vec.reshape((dens_vec.shape[0], 1)), (1, dens_vec.shape[0]))
    comp_2 = np.tile(dens_vec.reshape((1, dens_vec.shape[0])), (dens_vec.shape[0], 1))
    mask = comp_1 < comp_2
    return np.argmin(np.multiply(dist_mat, mask) + (np.multiply(np.logical_not(mask), np.max(dist_mat, axis=1))), axis=1)

def assign_to_cluster_dist(data, centroids):
    fact_1 = np.tile(data.reshape((data.shape[0],1,data.shape[1])), (1,centroids.shape[0],1))
    fact_2 = np.tile(centroids, (data.shape[0],1,1))
    return np.argmin(np.sum(np.square(fact_1 - fact_2), axis=2), axis=1)

def assign_to_cluster_nn(data, hdnn, centroid_idx):
    assigns = np.tile(np.array([-1,]), (data.shape[0])) # init assigns
    for c in range(centroid_idx.shape[0]):
        assigns[centroid_idx[c]] = c                    # fill centroids
    for i in range(data.shape[0]):
        if assigns[i] != -1:                            # if already assigned skip
            continue
        k = hdnn[i]                                    
        ks = [i,]
        while(assigns[k]) == -1:                        # loop up to the first neighbor filled
            ks.append(k)
            k = hdnn[k]
        kc = assigns[k]
        for k in ks:                                    # fill all the ones up to the one found
            assigns[k] = kc      
    return assigns
```

```python
# density peaks
def density_peaks_plot(data, dc=2.5):
    dm = distance_matrix(data)
    dv = density_vector(dm, dc)
    rv = radius_vector(dm, dv)
    plt.title("density peaks")
    plt.xlabel(r"$\rho$")
    plt.ylabel(r"$\delta$")
    plt.scatter(dv, rv)
    plt.show()
    
def density_peaks_clustering(data, n_clusters, dc=2.5):
    dm = distance_matrix(data)
    dv = density_vector(dm, dc)
    rv = radius_vector(dm, dv)
    sorted_gamma_idx = np.argsort(np.multiply(dv,rv))[::-1] # get idx of biggest values of gamma_i=delta_i*rho_i
    centroids = data[sorted_gamma_idx[:n_clusters]]         # get number of centroids
    hdnn = higher_dens_nn(dm, dv)                           # get list of higher density neighs
    return centroids, assign_to_cluster_nn(data, hdnn, sorted_gamma_idx[:n_clusters])

def plot_clusters_and_halo(data, assigns, centroids):
    for i in range(centroids.shape[0]):
        cluster = data[assigns==i]
        plt.scatter(cluster[:,0], cluster[:,1], s=4)
    halo = data[assigns==-1]
    plt.scatter(halo[:,0], halo[:,1], s=2)
    plt.scatter(centroids[:,0], centroids[:,1], s=12, c='k')
```

```python
# density peaks for aggregation dataset
density_peaks_plot(data)
```

```python
# compute clusters based on density peaks plot -> take 7 clusters (number of outliers above)
centroids, assigns = density_peaks_clustering(data, 7)
plot_clusters_and_halo(data, assigns, centroids)
```

#### Application to s3 dataset

```python
s3data = np.loadtxt("data/s3.txt")
```

```python
density_peaks_plot(s3data, dc=57500)
```

```python
# compute clusters based on plot -> take 15 clusters
centroids3, assigns3 = density_peaks_clustering(s3data, 15, 57500)
plot_clusters_and_halo(s3data, assigns3, centroids3)
```
